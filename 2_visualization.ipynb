```python
import pandas as pd

df = pd.read_parquet("flickr_data_cleaned.parquet")
```

```python
import folium
from folium.plugins import MarkerCluster

m = folium.Map(
    location=[df["lat"].median(), df["long"].median()],
    zoom_start=12,
    tiles="CartoDB positron"
)

cluster = MarkerCluster().add_to(m)

sample = df.sample(n=min(20000, len(df)), random_state=0)

for _, r in sample.iterrows():
    folium.Marker(
        location=[r["lat"], r["long"]],
        # Affichage de la photo quand on clique sur le marqueur (popup)
        
        popup=folium.Popup(
            f"""<b>{r["title"]}</b><br>
                <a href="{r["url"]}">{r["url"]}</a><br>""",
            max_width=300
        ),
        
    ).add_to(cluster)


```
